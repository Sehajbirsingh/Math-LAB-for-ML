The process below shows how to implement a simple linear regression analysis for a vehicle pricing scenario. Re-run it and take note of the R-sqaured value. Then. change the std value in np.random.normal to 1, 5 and 10. Observe the change in R-sqaured. How do you explain the change in R-sqaured value?

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

vehicle_age = np.random.uniform(0,10,1000)
vehicle_price = 70000 - (vehicle_age + np.random.normal(0, 10, 1000)) * 3000

vehicle_age = vehicle_age.reshape(-1, 1)  #reshape for single feature, use (1,-1) for single sample
vehicle_price = vehicle_price.reshape(-1, 1)

plt.scatter(vehicle_age, vehicle_price)
```

```python
reg = LinearRegression().fit(vehicle_age, vehicle_price)
reg.score(vehicle_age, vehicle_price)
```

R-squared measures how well the model explains the variability of the data. It is the proportion of variance in the dependent variable that is predictable from the independent variable.

When you increase the standard deviation of the noise (i.e., the std parameter in np.random.normal), the amount of random noise added to the vehicle price increases. This noise makes the data more scattered around the regression line, reducing the model's ability to explain the variance in the vehicle prices.

As a result, the R-squared value decreases with higher standard deviations of the noise. This is because the linear model fits the data less well when the data points are more spread out due to increased noise.
